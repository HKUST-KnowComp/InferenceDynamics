```python
import json
import pandas as pd
```

```python
with open('/home/hshiah/LLM_index/thirdpartyeval/plan-bench/responses/blocksworld/gpt-4_chat/task_1_plan_generation.json', 'r') as f:
    data = json.load(f)

data
```

```python
instances_dict = data['instances']
output_instance_dict = []
for instance in instances_dict:
    temp_dict = {}
    temp_dict['query'] = instance['query']
    temp_dict['llm_raw_response'] = instance['llm_raw_response']
    temp_dict['instance_id'] = instance['instance_id']
    temp_dict['example_instance_id'] = instance['example_instance_id']
    temp_dict['ground_truth_plan'] = instance['ground_truth_plan']
    output_instance_dict.append(temp_dict)
output_dict = {}
output_dict['instances'] = output_instance_dict
output_dict['task'] = data['task']
output_dict['engine'] = data['engine']
output_dict['prompt_type'] = data['prompt_type']
output_dict['domain'] = data['domain']
with open('/home/hshiah/LLM_index/thirdpartyeval/plan-bench/responses/blocksworld/gpt-4_chat/task_1_plan_generation.json', 'w') as f:
    json.dump(output_dict, f)
```
